```python
titanic.head(10)
```

```python
# for non numeric values pandas sorts by ABCD...Zabcd...z, which can make the sorting incorrect
titanic.sort_values('name')
```

```python
# the below function can solve this problem; it compares each string in lower case without changing the original dataset
titanic.sort_values('name', key=lambda col: col.str.lower())
```
